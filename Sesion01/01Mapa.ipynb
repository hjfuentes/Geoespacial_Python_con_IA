```python
!pip install folium -q
import folium

mapa = folium.Map(
    location=[4.142, -73.626],
    zoom_start=12,
    tiles="CartoDB positron")

ciudades = {"Villavicencio":[4.142,-73.626],
            "Acacías":[3.987,-73.758],
            "Cumaral":[4.270,-73.486]}

for n,c in ciudades.items():
    folium.Marker(c,popup=n).add_to(mapa)
mapa
```
